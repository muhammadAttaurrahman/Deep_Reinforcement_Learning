```python
episodes = 5 # Episodes are used for, how many times you want to check your environment
# # Episode can be a check an enverinment fully once

for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    scores = 0
    while not done:
        env.render()
        actions = env.action_space.sample()
        n_state, reward, done, info = env.step(actions)
        scores += reward
        print("Episode no: {} , Score: {}".format(episode,scores))
env.close()
```

## Train RL Model

```python
# Making the directries first 
log_path = os.path.join("logs")
log_path
```

```python
env_name = "CartPole-v0"
env = gym.make(env_name)
env = DummyVecEnv([lambda : env])
model = PPO ("MlpPolicy", env , verbose = 1, tensorboard_log = log_path) 
# Policy means how o operate in particular environment
# verbose means we want to log out the result for that particular environment
```

```python
# to get information about an algorithm 
# PPO??
```

```python
# Trainning the model
model.learn(total_timesteps = 20000)
```

## Adding a calll back a to trainning stage 

```python
# # Optuna, hyperparamter tunning by Optuna in stable_basline
# # call back is a very important thing is yyou add a specific reward threshols 
# wich mean that when the model hits that specific threshold it will stop and in this way our model will
# not be de_stabalized or unstable helfull when we are dealing with a large dataset/models
saved_path = os.path.join("Saved Models")
saved_path
```

 Training strategies 
1. Train for long time 
2. Tuning hypermetere
3. Try different algorithm

```python
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold

stop_callback = StopTrainingOnRewardThreshold(reward_threshold = 200, verbose= 1)

eval_callback = EvalCallback( env,
                            callback_on_new_best = stop_callback,
                             eval_freq= 10000,
                             best_model_save_path= saved_path,
                             verbose = 1)
```

```python
model = PPO("MlpPolicy", env , verbose = 1, tensorboard_log = log_path)
```

```python
model.learn(total_timesteps = 20000, callback = eval_callback)
```

## Changing policies 

```python
# We change policies in order to apply a new nural network to the algorithm to train our model 
net_arch = [dict(pi=[128,128,128,128], vf=[128,128,128,128])] # 128 represents units in 4 layers of the network
# pi is for custom actor
# vf is for value fumction
model = PPO("MlpPolicy", env , verbose = 1, tensorboard_log = log_path, policy_kwargs= {"net_arch":net_arch})
model.learn(total_timesteps = 20000, callback = eval_callback)
```

## Using an alternate algorithm 

```python
from stable_baselines3 import DQN

model = DQN("MlpPolicy", env , verbose = 1, tensorboard_log = log_path)
model.learn(total_timesteps = 20000, callback = eval_callback)
```
