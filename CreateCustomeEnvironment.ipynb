```python
#importing gym stuff
import gym
from gym import  Env
from gym.spaces import Box, Discrete, Dict, Tuple, MultiBinary, MultiDiscrete

#Importing Helpers
import numpy as np
import random
import os 

# Importing STableBaslines stuff
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
```

## Building an environment



```python
#Builiding an agent to give us the best shower possible 
#2_randomly temprature 
#37 to 39

class showerEnv(Env):
    def __init__(self):
        self.action_space =  Discrete(3)
        self.observation_space = Box(low=0, high= 100, shape=(1, ))
        self.state = 38 + random.randint(-3,3)
        self.shower_length = 60
                            
        pass
    def step(self, actions):
        # Applying the temp Adjacent
        self.state += actions -1 
        # decrease shower time
        self.shower_length -=1
#         self.observation_space -= 1
        # Calculating award 
        if self.state >=37 and self.state <=39:
            reward = 1
        
        else:
            reward = -1
        
        if (self.shower_length <=0):
            done = True
        
        else:
            done= False    
        info = {}
        return self.state, reward, done , info
        pass
    def render(self):
        pass
    def reset(self):
        self.state= np.array ([38 + random.randint(-3,3)]).astype(float)
        self.shower_length = 60
        return self.state
        pass
```

```python
env= showerEnv()
```

## Test The environment

```python
episodes = 5 # Episodes are used for, how many times you want to check your environment
# # Episode can be a check an enverinment fully once

for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    scores = 0
    while not done:
        env.render()
        actions = env.action_space.sample()
        n_state, reward, done, info = env.step(actions)
        scores += reward
        print("Episode no: {} , Score: {}".format(episode,scores))
env.close()
```

## Train the model

```python
log_path = os.path.join("CustomeEnv","logs")
model = PPO ("MlpPolicy", env , verbose = 1, tensorboard_log = log_path)
```

```python
model.learn(total_timesteps=500000)
```
