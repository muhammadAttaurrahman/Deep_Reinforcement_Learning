```python
import gym
import os 
from stable_baselines import ACER 
from stable_baselines.common.vec_env import DummyVecEnv
from stable_baselines.common.evaluation import evaluate_policy
```

```python
environment_name = "LunarLander-v2"
env = gym.make(environment_name)

env.observation_space
```

```python
env.action_space
```

```python
episodes = 5
n_state = 0 

for episodes in range(1, episodes+1):
    state = env.reset()
    done = False
    scores = 0 
    while not done:
        env.render()
        actions = env.action_space.sample()
        n_state, reward, done, info= env.step(actions)
        scores +=reward
        print(" Episode {} Score : {}" .format(episodes, scores))

env.close()
```

```python
# Training the agent 
env = gym.make(environment_name)
env = DummyVecEnv([lambda : env])
log_path = os.path.join('Training','logs')
model = ACER("MlpPolicy", env, verbose = 1 , tensorboard_log= log_path)
```

```python
model.learn(total_timesteps = 200000)
```

```python
evaluate_policy(model, env, n_eval_episodes=10, render = True)
```

```python
env.close()
```

```python
ppo_path = os.path.join ('Training','Saved Model', "ACER_LunarLander")
model.save(ppo_path)
```

```python
obs = env.reset()
while True :
    actions, _state = model.predict(obs)
    obs, reward, done, info = env.step(actions)
    env.render()
```
